```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------- Task 1: Hourly Call Simulation -------------------
print("=== Task 1: Simulating Hourly Call Volume ===")

# Parameters
lam = 15       # average number of calls per hour
hours = 8      # working day (8 hours)

# Generate Poisson-distributed random values
calls_per_hour = np.random.poisson(lam, hours)

# Compute summary statistics
mean_calls = np.mean(calls_per_hour)
var_calls = np.var(calls_per_hour)
std_calls = np.std(calls_per_hour)

print(f"λ (Expected mean calls/hour): {lam}")
print(f"Simulated Mean: {mean_calls:.2f}")
print(f"Variance: {var_calls:.2f}")
print(f"Standard Deviation: {std_calls:.2f}")
print(f"Hourly Call Counts: {calls_per_hour.tolist()}")
print()

# Plot histogram of hourly calls
plt.figure(figsize=(10,5))
plt.hist(calls_per_hour, bins=range(min(calls_per_hour), max(calls_per_hour)+2),
         color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(lam, color='red', linestyle='--', label=f'Expected mean = {lam}')
plt.title("Histogram: Hourly Customer Calls (λ = 15)")
plt.xlabel("Number of Calls per Hour")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Interpretation
print("Interpretation:")
print("------------------------------------------------")
print(f"- The simulated mean ≈ {mean_calls:.1f}, close to λ = 15.")
print("- Poisson distribution shows moderate variability around the mean.")
print("- The histogram is slightly right-skewed (common for Poisson).")
print("- Overall, call volumes are somewhat consistent hour to hour,")
print("  indicating predictable staffing needs with some variation.\n")


# ------------------- Task 2: Effect of λ on Distribution -------------------
print("=== Task 2: Impact of Mean (λ) on Distribution Shape ===")

lambda_values = [5, 10, 15, 25, 40]
size = 1000  # number of simulated hours for each λ

plt.figure(figsize=(12, 8))
for i, lam in enumerate(lambda_values):
    data = np.random.poisson(lam, size)
    plt.subplot(3, 2, i+1)
    plt.hist(data, bins=range(0, max(data)+2),
             color='orange', edgecolor='black', alpha=0.7)
    plt.axvline(lam, color='red', linestyle='--', label=f'Mean = {lam}')
    plt.title(f"λ = {lam}")
    plt.xlabel("Calls per Hour")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(alpha=0.25)

plt.tight_layout()
plt.show()

# Interpretation of λ effects
print("Interpretation:")
print("------------------------------------------------")
print("- For λ = 5: Distribution is highly right-skewed and discrete.")
print("- As λ increases to 10 and 15: The spread increases but shape starts to smooth.")
print("- For λ = 25 and 40: Distribution becomes more symmetric and bell-shaped,")
print("  resembling a normal distribution.")
print("- In practice, higher λ means more consistent hourly call volumes —")
print("  the relative randomness decreases (Law of Large Numbers).")
print("- Hence, high-traffic shifts (e.g., λ=40) show less fluctuation proportionally,")
print("  allowing management to plan staffing with greater confidence.\n")
```
