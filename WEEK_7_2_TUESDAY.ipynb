```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Parameters
n = 100   # number of deliveries
p = 0.6   # probability of on-time delivery

# Simulate outcomes for 100 deliveries
simulated_result = np.random.binomial(n, p)
print(f"Number of on-time deliveries (one trial of 100 deliveries): {simulated_result}")

# Multiple simulations (10,000 festive seasons)
simulated_trials = np.random.binomial(n, p, 10000)
print(f"Average on-time deliveries across simulations: {np.mean(simulated_trials):.2f}")
```

```python
# Possible outcomes: 0 to n deliveries on time
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Mean and variance
mean = n * p
variance = n * p * (1 - p)

# Plot PMF
plt.bar(x, pmf, color="skyblue", edgecolor="black")
plt.axvline(mean, color="red", linestyle="--", label=f"Mean = {mean}")
plt.title(f"Binomial PMF (n={n}, p={p})")
plt.xlabel("Number of On-Time Deliveries")
plt.ylabel("Probability")
plt.legend()
plt.show()

print(f"Mean: {mean}, Variance: {variance}")
```

```python
probabilities = [0.3, 0.6, 0.9]
plt.figure(figsize=(12, 8))

for i, prob in enumerate(probabilities, 1):
    pmf = binom.pmf(x, n, prob)
    mean = n * prob
    variance = n * prob * (1 - prob)

    plt.subplot(3, 1, i)
    plt.bar(x, pmf, color="lightgreen", edgecolor="black")
    plt.axvline(mean, color="red", linestyle="--", label=f"Mean={mean:.1f}, Var={variance:.1f}")
    plt.title(f"Binomial Distribution (n={n}, p={prob})")
    plt.xlabel("On-Time Deliveries")
    plt.ylabel("Probability")
    plt.legend()

plt.tight_layout()
plt.show()
```
