```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Parameters
n = 100   # number of trials
p = 0.3   # probability of success

# Simulate one experiment (100 queries)
simulated_result = np.random.binomial(n, p)
print(f"Number of successful resolutions in 100 queries: {simulated_result}")

# Simulate multiple experiments (e.g., 10,000 runs)
simulated_trials = np.random.binomial(n, p, 10000)
print(f"Average successes over simulations: {np.mean(simulated_trials):.2f}")
```

```python
# Define possible outcomes: 0 to n successes
x = np.arange(0, n+1)

# PMF using scipy
pmf = binom.pmf(x, n, p)

# Plot PMF
plt.bar(x, pmf, color="skyblue", edgecolor="black")
plt.title(f"Binomial PMF (n={n}, p={p})")
plt.xlabel("Number of Successful Resolutions")
plt.ylabel("Probability")
plt.show()

# Mean and variance
mean = n * p
variance = n * p * (1 - p)
print(f"Mean: {mean}, Variance: {variance}")
```

```python
probabilities = [0.1, 0.3, 0.5, 0.7]

plt.figure(figsize=(12,8))

for i, prob in enumerate(probabilities, 1):
    pmf = binom.pmf(x, n, prob)
    plt.subplot(2, 2, i)
    plt.bar(x, pmf, color="lightgreen", edgecolor="black")
    plt.title(f"Binomial PMF (n={n}, p={prob})")
    plt.xlabel("Successful Resolutions")
    plt.ylabel("Probability")

plt.tight_layout()
plt.show()
```
